```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# fetching datasets
df_1 = pd.read_csv('/content/drive/MyDrive/House Prices - Advanced Regression Techniques/train.csv')
df_2 = pd.read_csv('/content/drive/MyDrive/House Prices - Advanced Regression Techniques/test.csv')
```

```python
df_1.info(), df_2.info()
```

```python
# add SalesPrice col in test set before merging with train set
df_2['SalePrice'] = -1

df = pd.concat([df_1, df_2], axis = 0)

df.info()
```

```python
# sns heatmap
# Create a heatmap to visualize NaN values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of NaN Values')
plt.xlabel('columns')
plt.show()
```

```python
df.drop(['Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'MiscFeature', 'Fence'], inplace = True, axis = 1)
```
